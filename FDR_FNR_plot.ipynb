```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
import numpy as np
```

```python
longcallR_FDR = np.array([4.189686924,
6.830985915,
17.7178749,
38.08268819])
longcallR_FNR = np.array([60.61695685,
41.92273924,
31.9333189,
28.5225328])
clair3_FDR = np.array([39.54167948,
37.73006135,
39.06700866,
58.57936276])
clair3_FNR = np.array([25.60560182,
19.79806848,
17.70946092,
19.16714204])

title = "Evaluation of SNP calls on dataset Hek293t-drna"

x_longcallR = [1,2,3,4]
x_clair3 = [1.3,2.3,3.3,4.3]
x_label = ["10-20x","20-40x","40-100x","100+x"]
```

```python
# fig, ax = plt.subplots()
fig, (ax0, ax1) = plt.subplots(2, 1)
plt.rcParams['figure.figsize']=(6.4, 3.5)
plt.suptitle(title)
fig.tight_layout()
# plt.tick_params(labelsize=14)

ax0.spines['right'].set_visible(False)
ax0.spines['top'].set_visible(False)
ax0.spines['bottom'].set_visible(False)
ax0.bar(x_longcallR,longcallR_FDR,width=0.2, color='#4067B9',zorder = 1)
ax0.bar(x_clair3,clair3_FDR,width=0.2, color='#E37234',zorder = 1)
ax0.get_xaxis().set_visible(False)
ax0.set_ylabel("FDR*100", fontdict={'size':14})
ax0.spines['left'].set_linewidth(2)
# ax0.legend(["ours","clair3",], loc="upper center",frameon=False, ncol=2,prop = {'size':14})
ax0.tick_params(axis='both',labelsize=12)
ax0.yaxis.set_major_locator(ticker.MultipleLocator(max(longcallR_FDR.max()/4, clair3_FDR.max()/4).round()))

ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.invert_yaxis()
ax1.bar(x_longcallR, longcallR_FNR,width=0.2, color='#4067B9',zorder = 1)
ax1.bar(x_clair3,clair3_FNR,width=0.2,color='#E37234',zorder = 1)
ax1.set_ylabel("FNR*100", fontdict={'size':14})
ax1.spines['left'].set_linewidth(2)
ax1.tick_params(axis='both',labelsize=12)
# ax1.yaxis.set_major_locator(plt. MaxNLocator(integer=True))
ax1.yaxis.set_major_locator(ticker.MultipleLocator(max(longcallR_FNR.max()/4, clair3_FNR.max()/4).round()))

plt.axhline(0, color='black',linestyle="dashed", linewidth=1, zorder=2)
plt.subplots_adjust(wspace=0, hspace=0)
plt.xticks([1.1,2.1,3.1,4.1],["10-20x","20-40x","40-100x","100+x"])
plt.xlabel("coverage", fontdict={"size":14})
fig.align_ylabels()
fig.legend(["longcallR","clair3"],frameon=False)
plt.savefig("figs/hg002_masseq_while_regions_FDR_FNR.svg", dpi=1200, format="svg", bbox_inches = 'tight')
```
